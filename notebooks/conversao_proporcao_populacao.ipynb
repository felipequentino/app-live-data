```python
import pandas as pd

file_path = '../data/data_predicted/predict_populacao_2024-2025.csv'
df = pd.read_csv(file_path)
df
```

```python
PROPORCAO_POPULACAO_HOMENS = 0.48863552
PROPORCAO_POPULACAO_MULHERES = 0.51136448
```

```python
df_h = df.copy()
df_m = df.copy()
df_h['populacao'] = round(df_h['populacao'] * PROPORCAO_POPULACAO_HOMENS)
df_m['populacao'] = round(df_m['populacao'] * PROPORCAO_POPULACAO_MULHERES)
df_m
```

```python
df_h.to_csv('predict_populacao_H_2024-2025.csv', index=False)
df_m.to_csv('predict_populacao_M_2024-2025.csv', index=False)
```
