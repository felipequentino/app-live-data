```python
import pandas as pd

file_path = '../data/data_predicted/predict_ocupacao_2024-2025.csv'
df = pd.read_csv(file_path)
df
```

```python
PROPORCAO_OCUPACAO_HOMENS = 0.573156595
PROPORCAO_OCUPACAO_MULHERES = 0.426843405
```

```python
df_h = df.copy()
df_m = df.copy()
df_h['populacao'] = round(df_h['populacao'] * PROPORCAO_OCUPACAO_HOMENS)
df_m['populacao'] = round(df_m['populacao'] * PROPORCAO_OCUPACAO_MULHERES)
df_m
```

```python
df_h.to_csv('predict_ocupacao_H_2024-2025.csv', index=False)
df_m.to_csv('predict_ocupacao_M_2024-2025.csv', index=False)
```
