```python
import pandas as pd

file_path = '../data/data_predicted/predict_desocupados_2024-2025.csv'
df = pd.read_csv(file_path)
df
```

```python
PROPORCAO_DESOCUPACAO_HOMENS = 0.4832
PROPORCAO_DESOCUPACAO_MULHERES = 0.5167
```

```python
df_h = df.copy()
df_m = df.copy()
df_h['populacao'] = round(df_h['populacao'] * PROPORCAO_DESOCUPACAO_HOMENS)
df_m['populacao'] = round(df_m['populacao'] * PROPORCAO_DESOCUPACAO_MULHERES)
df_m
```

```python
df_h.to_csv('predict_desocupados_H_2024-2025.csv', index=False)
df_m.to_csv('predict_desocupados_M_2024-2025.csv', index=False)
```
