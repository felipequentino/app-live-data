```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

from statsmodels.tsa.stattools import adfuller 
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX
from pandas.tseries.offsets import DateOffset
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf


from statsmodels.tsa.arima.model import ARIMA
```

```python
df = pd.read_csv("../data/economico/veiculos_2016-2023.csv")
df
```

```python
idx = pd.period_range(start='2016-01', end='2024-01', freq='M')
idx
```

```python
lt = df["LicenciamentoTotal"].values
ln = df["LicenciamentoNacionais"].values
li = df["LicenciamentoImportados"].values
pr = df["Produção"].values
ex = df["Exportação"].values
```

```python
df_licenciamento_total = pd.DataFrame(data={"populacao":lt}, index=idx)
df_licenciamento_nacionais = pd.DataFrame(data={"populacao":ln}, index=idx)
df_licenciamento_importados = pd.DataFrame(data={"populacao":li}, index=idx)
df_pr = pd.DataFrame(data={"populacao":pr}, index=idx)
df_ex = pd.DataFrame(data={"populacao":ex}, index=idx)

df_ex
```

```python
df_licenciamento_total.to_csv("../data/veiculos/licenciamento_total_2016-2023.csv")
df_licenciamento_nacionais.to_csv("../data/veiculos/licenciamento_nacionais_2016-2023.csv")
df_licenciamento_importados.to_csv("../data/veiculos/licenciamento_importados_2016-2023.csv")
df_pr.to_csv("../data/veiculos/producao_2016-2023.csv")
df_ex.to_csv("../data/veiculos/exportacao_2016-2023.csv")
```
